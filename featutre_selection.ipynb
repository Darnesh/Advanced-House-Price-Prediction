```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

pd.set_option('display.max_columns',None)
```

```python
data = pd.read_csv("/home/darnesh/Documents/programming/Machine Learning/house-prices-advanced-regression-techniques/featured_train.csv")
data.head()
```

```python
x_train = data.drop(['Id','SalePrice'],axis=1)
y_train = data['SalePrice']
```

```python
test = pd.read_csv("/home/darnesh/Documents/programming/Machine Learning/house-prices-advanced-regression-techniques/featured_test.csv")
x_test = test.drop('Id',axis=1)
```

```python
x_test['GarageCars'].isnull().any()
```

```python
x_test['GarageCars'] = x_test['GarageCars'].median()
```

```python
# bigger the alpha the less features will be selected
lasso = Lasso(alpha=0.05,random_state=42)

# SelectFromModel selects a subset of features based on importance scores or thresholds derived from a specific machine learning model.
# selects the feature with a non-zero coefficient
select_model = SelectFromModel(estimator=lasso)
select_model.fit(x_train,y_train)
```

```python
select_model.get_support()
```

```python
sel_columns = x_train.columns[(select_model.get_support())]
sel_columns
```

```python
x_train = x_train[sel_columns]
x_train.head()
```

```python
test = test[sel_columns]
```

```python
lasso.fit(x_train,y_train)
```

```python
pred = lasso.predict(test)
```

```python
pred
```

```python
import numpy as np
test['SalePrice'] = np.round(np.exp(pred),0)

test.sample(10)
```
